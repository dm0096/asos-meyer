```python
"""
Plotting ASOS netCDF4_CLASSIC data.

Dean Meyer 2021
"""

import hvplot.xarray
import hvplot.pandas
import numpy as np
import xarray as xr
import pandas as pd
import panel as pn

# file path
file = r"/nas/rstor/dmeyer/GRA/ASOS/KHSV201711.nc"

# open an xarray dataset
ds = xr.open_dataset(
    file, decode_timedelta=False
)  # dont decode times from s into ns(!!)

# plotting via xarray
# print(ds.variables)

# read dataset into DataFrame
df = ds.to_dataframe()
df["time"] = pd.to_datetime(df["epochTime"], unit="s")

df['ID'] = 'KHSV'

df = df.set_index(["ID", "time"], drop=True)

# do all that again for another file
file2 = r"/nas/rstor/dmeyer/GRA/ASOS/KDCU201711.nc"

ds2 = xr.open_dataset(
    file, decode_timedelta=False
)  # dont decode times from s into ns(!!)

df2 = ds2.to_dataframe()
df2["time"] = pd.to_datetime(df2["epochTime"], unit="s")

df2['ID'] = 'KDCU'

df2 = df2.set_index(["ID", "time"], drop=True)

# do all that again for another file
file3 = r"/nas/rstor/dmeyer/GRA/ASOS/KMSL201711.nc"

ds3 = xr.open_dataset(
    file, decode_timedelta=False
)  # dont decode times from s into ns(!!)

df3 = ds3.to_dataframe()
df3["time"] = pd.to_datetime(df3["epochTime"], unit="s")

df3['ID'] = 'KMSL'

df3 = df3.set_index(["ID", "time"], drop=True)

# combine DataFrames
df = pd.concat([df, df2, df3])

# print(df.head(-10))

# df.loc['KMSL']

# Clearing out missing values
# print(df[df['Dewpoint Temperature'] == -2147483648])
df['Dewpoint Temperature'] = df['Dewpoint Temperature'][df['Dewpoint Temperature'] > -200]
```

```python
# plotting
y = pn.widgets.Select(name='y', options=list(df.columns))
kind = pn.widgets.Select(name='kind', value='scatter', options=['line', 'scatter'])
plot = df.loc['KMSL'].hvplot(y=y, kind=kind, colorbar=False, width=600, height=400)
pn.Row(pn.WidgetBox(y, kind), plot)
```
